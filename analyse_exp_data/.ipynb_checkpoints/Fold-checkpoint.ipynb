```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Conditions
conditions = ["0h", "2.5h ctrl", "2.5h agar", "5h ctrl", "5h agar"]

# Dictionnaire pour stocker les valeurs pour chaque condition
values_fold = {
    "0h": [0.201995718,0, 0.088662922],  
    "2.5h ctrl": [0.084384846, 0.09194583, 0.286093629, 0.153685956],
    "2.5h agar": [0.563647783, 0.113057776, 0.193722532, 0.205122608],
    "5h ctrl": [0.303412787, 0.144926757, 0.162371617],
    "5h agar": [0.142683246, 0.139429177, 0.130015471], 
}
max_length = max(len(values) for values in values_fold.values())
fold_data = pd.DataFrame({key: values + [None] * (max_length - len(values)) for key, values in values_fold.items()})
fold_data.to_csv('fold_data', index=True)
```

```python
# Conditions et moyennes
conditions = ["0h", "2.5h ctrl", "2.5h agar", "5h ctrl", "5h agar"]
means = [0.09632932, 0.154527065, 0.268637425, 0.20357072033333333, 0.137709298]

# Tracé du graphique
plt.figure(figsize=(10, 6))
plt.plot(conditions, means, marker='o', color='skyblue', linestyle='-')
plt.title('Moyenne des données par condition')
plt.xlabel('Condition')
plt.ylabel('Moyenne')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.boxplot(values_fold.values())
plt.xlabel('Condition')
plt.ylabel('Fold')
plt.xticks(range(1, len(values_fold) + 1), values_fold.keys())  # Utilisation des clés pour les étiquettes sur l'axe x
plt.grid(True)
plt.ylim(0,0.4)
plt.savefig('plot_aniso_constraint1.png', transparent=False)  # Le chemin du fichier de sauvegarde semble être incorrect
plt.show()
```
