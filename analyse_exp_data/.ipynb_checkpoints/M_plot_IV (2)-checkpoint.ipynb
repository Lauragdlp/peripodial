```python
from scipy.stats import shapiro, ttest_ind
import statsmodels.api as sm
# Fonction pour effectuer les tests de normalité et t-tests
def perform_tests(ctrl_data, agar_data, y_pos, star_pos):
    shapiro_ctrl = shapiro([x for x in ctrl_data if x is not None])
    shapiro_agar = shapiro([x for x in agar_data if x is not None])

    print(f"Test de Shapiro-Wilk pour ctrl : {shapiro_ctrl}")
    print(f"Test de Shapiro-Wilk pour agar : {shapiro_agar}")

    t_test_result = ttest_ind([x for x in ctrl_data if x is not None],
                              [x for x in agar_data if x is not None],
                              nan_policy='omit')

    print(f"Test t de Student entre ctrl et agar : {t_test_result}")

    if t_test_result.pvalue < 0.05:
        plt.text(star_pos, y_pos, '***', fontsize=12)
    elif t_test_result.pvalue < 0.01:
        plt.text(star_pos, y_pos, '****', fontsize=12)
```

```python
def perform_tests_ct(ctrl_data1, ctrl_data2, y_pos, star_pos):
    # Filtrer les données None
    ctrl_data1 = [x for x in ctrl_data1 if x is not None]
    ctrl_data2 = [x for x in ctrl_data2 if x is not None]

    # Test de normalité de Shapiro-Wilk
    shapiro_ctrl1 = shapiro(ctrl_data1)
    shapiro_ctrl2 = shapiro(ctrl_data2)

    print(f"Test de Shapiro-Wilk pour ctrl1 : {shapiro_ctrl1}")
    print(f"Test de Shapiro-Wilk pour ctrl2 : {shapiro_ctrl2}")

    # Test t de Student
    t_test_result = ttest_ind(ctrl_data1, ctrl_data2, nan_policy='omit')

    print(f"Test t de Student entre ctrl1 et ctrl2 : {t_test_result}")

    # Afficher les étoiles pour les p-values significatives
    if t_test_result.pvalue < 0.01:
        plt.text(star_pos, y_pos, '****', fontsize=12)
    elif t_test_result.pvalue < 0.05:
        plt.text(star_pos, y_pos, '***', fontsize=12)
```
