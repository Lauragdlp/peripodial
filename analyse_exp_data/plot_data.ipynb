```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Load the DataFrame from the CSV file
F3_WP = pd.read_csv('../experimental_data/F3_WP.csv')
```

```python
conditions = ["0h", "2.5h ctrl", "2.5h agar", "5h ctrl", "5h agar"]
values_areas = {
    "0h": [],  
    "2.5h ctrl": [],
    "2.5h agar": [],
    "5h ctrl": [],
    "5h agar": [], 
}

values_areas["0h"].extend(F3_WP.area.to_list())
```

```python
for condition in conditions:
    plt.scatter([condition] * len(values_areas[condition]), values_areas[condition], label=condition)

# Ajouter des lÃ©gendes et des titres
plt.xlabel("Condition")
plt.ylabel("Area")
plt.title("Aire des cellules en fonction de la contrainte et du developpement")
plt.legend()
plt.savefig('../analyse_exp_data/Figures/Area_cell_IV.png', transparent=True)
# Afficher le nuage de points
plt.show()
```

```python
values_aniso = {
    "0h": [],  
    "2.5h ctrl": [],
    "2.5h agar": [],
    "5h ctrl": [],
    "5h agar": [], 
}

values_aniso["0h"].extend(F3_WP.aniso)
```

```python
for condition in conditions:
    plt.scatter([condition] * len(values_aniso[condition]), values_aniso[condition], label=condition)

# Ajouter des lÃ©gendes et des titres
plt.xlabel("Condition")
plt.ylabel("Anisotropie")
plt.title("Anisotropie des cellules en fonction de la contrainte et du developement")
plt.legend()
plt.savefig('../analyse_exp_data/Figures/Aniso_cell_IV.png', transparent=True)
# Afficher le nuage de points
plt.show()
```

```python
values_perimeter = {
    "0h": [],  
    "2.5h ctrl": [],
    "2.5h agar": [],
    "5h ctrl": [],
    "5h agar": [], 
}

values_perimeter["0h"].extend(F3_WP.perimeter)
values_perimeter["2.5h ctrl"].extend(F3_WP.perimeter)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Supposons que conditions soit une liste de diffÃ©rentes conditions
# Supposons que values_aniso soit un dictionnaire contenant les valeurs d'anisotropie pour chaque condition

# CrÃ©ation de la figure
fig = plt.figure(figsize=(20, 10))

# Tracer les boÃ®tes Ã  moustaches pour chaque condition
sns.boxplot(data=[values_aniso[condition] for condition in conditions],
            palette="colorblind",  # Utilisation de la palette colorblind
            order=[1, 0.9, 0.8, 0.7, 0.6, 0.5, 0.4],
            showfliers=False)

# Ajouter des lÃ©gendes et des titres
plt.xlabel("Condition")
plt.ylabel("Anisotropie")
plt.title("Anisotropie des cellules en fonction de la contrainte")

# Afficher le graphique
plt.show()
```

```python
import seaborn as sns

plt.boxplot([values_perimeter[condition] for condition in conditions], labels=conditions, patch_artist=True)

fig = plt.figure(figsize=(20, 10))

# Tracer les boÃ®tes Ã  moustaches pour chaque condition
for condition in conditions:
    sns.boxplot(x=[condition] * len(values_aniso[condition]), y=values_aniso[condition], 
                palette="colorblind",
                showfliers=False)
# Ajouter des lÃ©gendes et des titres
plt.xlabel("Condition")
plt.ylabel("Perimeter")
plt.title("Â¨PerimÃ¨tre des cellules en fonction de la contrainte et du developement")
plt.legend()
plt.savefig('../analyse_exp_data/Figures/Peri_cell_IV.png', transparent=True)

# Afficher le nuage de points
plt.show()
```
