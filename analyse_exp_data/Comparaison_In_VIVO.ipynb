```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Load the DataFrame from the CSV file
F3_WP = pd.read_csv('../experimental_data/F3_WP.csv')
```

```python
conditions_bis = ["Sans PE", "0h PE", "PE + agar 2.5h", "PE + 5h agar"]
values_areas_bis = {
    "Sans PE": [],  
    "0h PE": [],
    "PE + agar 2.5h": [],
    "PE + 5h agar": [],
}

values_areas_bis["0h PE"].extend(F3_WP.area.to_list())
```

```python
for condition in conditions_bis:
    plt.scatter([condition] * len(values_areas_bis[condition]), values_areas_bis[condition], label=condition)

plt.xlabel("Condition")
plt.ylabel("Area")
plt.title("Aire des cellules en fonction de la contrainte et du developpement")
plt.legend()
plt.savefig('../analyse_exp_data/Figures/Compa_Area_cell_IV.png', transparent=True)
plt.show()
```

```python
values_aniso_bis = {
    "Sans PE": [],  
    "0h PE": [],
    "PE + agar 2.5h": [],
    "PE + 5h agar": [],
}

values_aniso_bis["0h PE"].extend(F3_WP.aniso)
```

```python
for condition in conditions_bis:
    plt.scatter([condition] * len(values_aniso_bis[condition]), values_aniso_bis[condition], label=condition)

plt.xlabel("Condition")
plt.ylabel("Anisotropie")
plt.title("Anisotropie des cellules en fonction de la contrainte et du developement")
plt.legend()
plt.savefig('../analyse_exp_data/Figures/Compa_Aniso_cell_IV.png', transparent=True)
plt.show()
```

```python
values_perimeters_bis = {
    "Sans PE": [],  
    "0h PE": [],
    "PE + agar 2.5h": [],
    "PE + 5h agar": [],
}

values_perimeters_bis["0h PE"].extend(F3_WP.aniso)
```

```python
for condition in conditions_bis:
    plt.scatter([condition] * len(values_perimeters_bis[condition]), values_perimeters_bis[condition], label=condition)

plt.xlabel("Condition")
plt.ylabel("Perimeter")
plt.title("Perimtre des cellules en fonction de la contrainte et du developement")
plt.legend()
plt.savefig('../analyse_exp_data/Figures/Compa_Peri_cell_IV.png', transparent=True)

plt.show()
```
