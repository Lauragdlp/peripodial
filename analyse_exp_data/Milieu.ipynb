```python
import pandas as pd
import numpy as np

# Charger les données en spécifiant le séparateur correct
I_F20_5h_agar = pd.read_csv('../experimental_data/I_F20_5h_agar.csv', sep=';')


# Calculer la plage des valeurs de 'fx'
value = I_F20_5h_agar['fx'].max() - I_F20_5h_agar['fx'].min()

# Déterminer les indices des lignes à sélectionner
start_value = value / 3
end_value = value * 2 / 3

# Trouver les indices des valeurs les plus proches de start_value et end_value
start_index = (I_F20_5h_agar['fx'] - start_value).abs().idxmin()
end_index = (I_F20_5h_agar['fx'] - end_value).abs().idxmin()

# Assurez-vous que les index ne dépassent pas les limites du DataFrame
start_index = max(0, start_index)
end_index = min(len(I_F20_5h_agar), end_index)

# Sélectionner les lignes correspondantes
filtered_df = I_F20_5h_agar.iloc[start_index:end_index]

# Sélectionner les colonnes 'aniso', 'area', et 'perimeter'
result = filtered_df

# Sauvegarder le DataFrame filtré dans un fichier CSV
result.to_csv('M_F20_5h_agar.csv', index=True)
```
